```python
filename='flipkart_products_new.csv'
f=open(filename,'w')
```

```python
headers='Title,Price,Ratings\n'
f.write(headers)
```

```python
for page in range(1,10):
    page_url=base_url+str(page)
    res=requests.get(page_url)
    soup=bs4.BeautifulSoup(res.text,'lxml')
    soup_version=soup.select("._3O0U0u")
    
    for item in soup_version:
        product_title=item.select('img')[0]['alt']
        final_product_title=product_title
        
        product_price=item.select(".col.col-5-12._2o7WAb")
        final_product_price=product_price[0].text.strip()
        
        product_rating=item.select(".niH0FQ")
        final_product_rating=product_rating[0].text
        
        trim_price=''.join(final_product_price.split(','))
        split_price=trim_price.split('₹')
        add_Rs='Rs.'+split_price[1]
        divide_price=add_Rs.split('N')
        final_price=divide_price[0]
        
        final_rating=final_product_rating[0:3]
        
        print(final_product_title.replace(',','|')+','+final_price+','+final_rating+"\n")
        f.write(final_product_title.replace(',','|')+','+final_price+','+final_rating+"\n")
f.close()
                                        
```
